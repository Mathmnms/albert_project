```python
import pandas as pd

# Assuming 'df' is your DataFrame containing the data

# Convert the '% Voix/Exp' column to numeric if it's not already
df['% Voix/Exp'] = pd.to_numeric(df['% Voix/Exp'], errors='coerce')

# Group by 'lib_commune' and 'year', then apply a function to get the row with the maximum '% Voix/Exp' for each group
result = df.loc[df.groupby(['lib_commune','lib_dep', 'year'])['% Voix/Exp'].idxmax()]

print(result)
```
